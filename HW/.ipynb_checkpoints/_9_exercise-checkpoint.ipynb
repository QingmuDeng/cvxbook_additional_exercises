```python
import numpy as np
import scipy as sp
import cvxpy as cp
import matplotlib.pyplot as plt
import jax, jax.numpy as jnp
```

## 9.30

```python
n = 100
m = 200

np.random.seed(0)

A = np.random.randn(m,n)
A.shape
```

```python
x = cp.Variable(n)

obj = -cp.sum(cp.log(1-A@x))-cp.sum(cp.log(1-x**2))

constraints = []
# constraints += [cp.max(A@x)<1]

prob = cp.Problem(
    cp.Minimize(obj),
    constraints
)

p_opt = prob.solve()
p_opt
```

```python
f = lambda x: -jnp.sum(jnp.log(1-A@x))-jnp.sum(jnp.log(1-x**2))
f_grad = jax.grad(f)
f_hess = jax.hessian(f)
```

### Gradient Descent

```python
x = np.zeros(n)

idx = 0

max_iter = 500

list_val = []
list_t = []

while idx < max_iter:
    α = 0.01
    β = 0.8 
    
    t = 1
    
    grad_f = f_grad(x)
    Δx = -grad_f
    
    while np.max(A@(x+t*Δx)) >= 1 or np.max(np.abs(x+t*Δx))>=1:
        t *= β
    
    while f(x+t*Δx) > f(x) + α*t*grad_f.T@Δx:
        t *= β

    
    
    x = x+t*Δx

    list_val.append(f(x))
    list_t.append(t)
    # print(f(x),np.linalg.norm(grad_f, 2))
    
    # Stopping Criterion
    if np.linalg.norm(grad_f, 2)<1e-3:
        break

    idx += 1
```

```python
fig, ax = plt.subplots()
ax.semilogy(abs(list_val-p_opt))
ax.set_title('f(x)-p*')
```

```python
fig, ax = plt.subplots()
ax.plot(list_t)
ax.set_title('Step Scaler')
```

### Newton's Method

```python
α = 0.3
β = 0.5

x = np.zeros(n)

idx = 0

max_iter = 100

list_val = []
list_t = []

while idx < max_iter:
    
    t = 1
    
    grad_f = f_grad(x)
    hess_f = f_hess(x)
    hess_f_inv = jnp.linalg.inv(hess_f)
    Δx_nt = -hess_f_inv@grad_f
    λ2 = grad_f.T@hess_f_inv@grad_f

    # Stopping Criterion
    if λ2/2<1e-3:
        break
    
    while np.max(A@(x+t*Δx_nt)) >= 1 or np.max(np.abs(x+t*Δx_nt))>=1:
        t *= β
    
    while f(x+t*Δx_nt) > f(x) + α*t*grad_f.T@Δx_nt:
        t *= β

    print(f(x),λ2)
    
    x = x+t*Δx_nt

    list_val.append(f(x))
    list_t.append(t)    
    

    idx += 1
```

```python
fig, ax = plt.subplots()
ax.semilogy(abs(list_val-p_opt),'o-')
ax.set_title('f(x)-p*')
```

```python
fig, ax = plt.subplots()
ax.plot(list_t,'o-')
ax.set_title('Step Scaler')
```

## Problem 9.31

### Reusing Hessian
Even updating hessian once every four steps works sufficiently well

```python
α = 0.3
β = 0.5

x = np.zeros(n)

idx = 0

max_iter = 100

list_val = []
list_t = []

while idx < max_iter:
    
    t = 1
    
    grad_f = f_grad(x)
    if idx%4==0:
        hess_f = f_hess(x)
        hess_f_inv = jnp.linalg.inv(hess_f)
    Δx_nt = -hess_f_inv@grad_f
    λ2 = grad_f.T@hess_f_inv@grad_f

    # Stopping Criterion
    if λ2/2<1e-3:
        break
    
    while np.max(A@(x+t*Δx_nt)) >= 1 or np.max(np.abs(x+t*Δx_nt))>=1:
        t *= β
    
    while f(x+t*Δx_nt) > f(x) + α*t*grad_f.T@Δx_nt:
        t *= β

    print(f(x),λ2)
    
    x = x+t*Δx_nt

    list_val.append(f(x))
    list_t.append(t)    
    

    idx += 1
```

```python
fig, ax = plt.subplots()
ax.semilogy(abs(list_val-p_opt),'o-')
ax.set_title('f(x)-p*')
```

```python
fig, ax = plt.subplots()
ax.plot(list_t,'o-')
ax.set_title('Step Scaler')
```

## A9.4 Efficient Numerical Methods for RLS

```python
n = 2000
k = 100
δ = 1
η = 1

np.random.seed(0)

A = np.random.randn(k,n)
b = np.random.randn(k,1)

I = np.eye(n,n)
Δ = -np.tri(n,n,1) + np.tri(n,n,-2) + 3*I
Δ[0,0] = 1 # first and last diagonal is 1
Δ[-1,-1] = 1 # first and last diagonal is 1
Δ
```

```python
%%time
x_opt = sp.linalg.solve(A.T@A+δ*Δ+η*I, A.T@b)
x_opt
```

```python
%%time
g = A.T@b

# Converting this banded matrix is a bit tricky!!
Ab = np.array([
    np.concat([[0], np.diag(δ*Δ+η*I,k=1)]),
    np.diag(δ*Δ+η*I,k=0),
    np.concat([np.diag(δ*Δ+η*I,k=-1),[0]]),
])
z1 = sp.linalg.solve_banded((1,1),Ab, g)
Z2 = sp.linalg.solve_banded((1,1),Ab, A.T)
z3 = sp.linalg.solve((np.eye(k)+A@Z2),A@z1)
x_opt = z1 - Z2@z3
x_opt
```
