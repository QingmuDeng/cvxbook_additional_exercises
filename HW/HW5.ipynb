### A4.18

```python
np.random.seed(0)
m, n = 300, 100
A = np.random.rand(m,n)
b = A.dot(np.ones(n)) / 2
c = -np.random.rand(n)
```

```python
x = cp.Variable(n)

obj = cp.Minimize(c.T@x)

constraints = []
constraints += [A@x<=b]
constraints += [x<=1]
constraints += [x>=0]

problem = cp.Problem(obj, constraints)

problem.solve()
```

```python
x.value
```

```python
fig, ax = plt.subplots(2,1)
array_obj_bl = []
array_max_vio = []
array_cons_vio = []
for t in np.linspace(0, 1, 100):
    x_bl = (x.value > t).astype(int)
    obj_bl = c.T@x_bl
    cons_vio = A@x_bl-b
    max_vio = np.max(cons_vio)
    cons_vio = any(cons_vio>=0)
    

    array_obj_bl.append(obj_bl)
    array_max_vio.append(max_vio)
    array_cons_vio.append(cons_vio)

array_obj_bl   = np.array(array_obj_bl)
array_max_vio = np.array(array_max_vio)
array_cons_vio = np.array(array_cons_vio)

ax[0].plot(np.linspace(0, 1, 100), array_obj_bl, '.:', label='feasible')
ax[0].plot(np.linspace(0, 1, 100)[array_cons_vio], array_obj_bl[array_cons_vio], '.:', label='infeasible')
ax[1].plot(np.linspace(0, 1, 100), array_max_vio, '.:', label='feasible')
ax[1].plot(np.linspace(0, 1, 100)[array_cons_vio], array_max_vio[array_cons_vio], '.:', label='infeasible')
ax[0].legend()
ax[1].legend()
ax[0].grid()
ax[1].grid()
ax[0].set_ylabel('Objective')
ax[1].set_ylabel('Max Constraint Violation')
ax[1].set_xlabel('t -- boolean conversion threshold')
```

The gap between the two values is 0.8399

### A4.42

```python
np.random.seed(0)
n = 10 
A = np.random.rand(n, n)
P = A.T@A
P
```

```python
x0 = np.random.rand(n)
list_x = [x0/np.linalg.norm(x0,2)]
list_obj = []
print('initial x:', list_x[0])
steps = 10
for i in range(steps):
    x_k = list_x[-1]

    x = cp.Variable(n)
    obj = cp.Minimize(-x_k.T@P@x_k-2*x_k.T@P@(x-x_k))

    constraints = [cp.norm2(x)<=1]

    problem = cp.Problem(obj, constraints)

    cost = problem.solve()

    list_obj.append(cost)

    list_x.append(x.value)

eig_val, eig_vec = np.linalg.eig(P)

fig, ax = plt.subplots()
ax.hlines(-eig_vec[:,0].T@P@eig_vec[:,0], 0, steps-0.5,'r',alpha=2/3, label='Optimal Value')
ax.plot(list_obj,'.--',alpha=2/3,label='Optimal Value per Iter')
ax.legend()
```

```python
k = 201
x = np.linspace(1, k, k)
t = -3 + 6*(x-1)/(k-1)
y = np.exp(t)


p = lambda t, a: a[0] + a[1]*t+ a[2]*t**2
q = lambda t, b: 1+b[0]*t+b[1]*t**2
y = lambda t: np.exp(t)
f = lambda t, a, b: p(t, a, b)/q(t, a, b)

a = cp.Variable(3)
b = cp.Variable(2)

γ = 0.01


obj = cp.Minimize(0)

constraints = []
constraints += [q(t[i], b)>=0 for i in range(k)]
constraints += [p(t[i],a)-q(t[i],b)*y(t[i])-γ*q(t[i],b)<=0 for i in range(k)]
constraints += [p(t[i],a)-q(t[i],b)*y(t[i])+γ*q(t[i],b)>=0 for i in range(k)]

prob = cp.Problem(obj, constraints)

prob.solve()#==np.inf
```

```python
l = 0
u = 0.1

while abs(l-u)>1e-4:
    γ = (l+u)/2

    obj = cp.Minimize(0)

    constraints = []
    constraints += [q(t[i], b)>=0 for i in range(k)]
    constraints += [p(t[i],a)-q(t[i],b)*y(t[i])-γ*q(t[i],b)<=0 for i in range(k)]
    constraints += [p(t[i],a)-q(t[i],b)*y(t[i])+γ*q(t[i],b)>=0 for i in range(k)]
    
    prob = cp.Problem(obj, constraints)
    
    result = prob.solve(solver='ECOS')
    print(result)

    if prob.status!='optimal': # result==np.inf
        l = γ
    else:
        u = γ


# A final round of solution with a feasible u
obj = cp.Minimize(0)

constraints = []
constraints += [q(t[i], b)>=0 for i in range(k)]
constraints += [p(t[i],a)-q(t[i],b)*y(t[i])-u*q(t[i],b)<=0 for i in range(k)]
constraints += [p(t[i],a)-q(t[i],b)*y(t[i])+u*q(t[i],b)>=0 for i in range(k)]

prob = cp.Problem(obj, constraints)

result = prob.solve(solver='ECOS')
```

## A6.5

```python
np.random.seed(0)
m = 100
n = 30
A = np.random.rand(m, n)
b = np.random.rand(m)

t = cp.Variable()
x = cp.Variable(n)

obj = cp.Minimize(t)

constraints = []
constraints += [cp.sum(
    cp.hstack(
        [cp.power(A[i,:]@x-b[i], 3/2) for i in range(m)]
    )
)<=t]

prob = cp.Problem(obj, constraints)

prob.solve()

plt.hist(A@x.value-b)
plt.title('$L_{1.5}$ Norm')
```

### A17.16

```python
r = 1.05
F = 0.9
C = 1.15
K1 = 1.1 # Call 1
K2 = 1.2 # Call 2
K3 = 0.8 # Put 1
K4 = 0.7 # Put 2


p0_5 = np.array([1, 1, 0.06, 0.03, 0.02, 0.01])
p6 = cp.Variable()
p = cp.hstack([p0_5, p6])

num = 200
S = np.linspace(0.5, 2, num)

V = np.array(
    [[r, Si, max(Si-K1,0), max(Si-K2,0), max(K3-Si,0), max(K4-Si,0), min(C,max(F,Si))] for Si in S]
    # r, Si,      call 1,     call 2,     put 1,        put 2,       collar        
)

y = cp.Variable(num)

obj = cp.Minimize(p6)

constraints = []
constraints += [-V.T@y+p==0]
constraints += [y>=0]

prob = cp.Problem(obj, constraints)

prob.solve()
```
